```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

## Step Function : 
A step activation function is a binary activation function that outputs one if the input is above a certain threshold and zero otherwise. 
\(f(x)=1\) if \(x\ge 0\) \(f(x)=0\) if \(x<0\)Â 

```python
def step(x):
    return np.where(x>=0,1,0)
```

```python
x = np.linspace(-5, 5, 50)
y = step(x)
```

```python
plt.plot(x, y, label='Step function', color='blue')
plt.title('Step Function')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

## Sigmoid Function:
A sigmoid function is any mathematical function whose graph has a characteristic S-shaped or sigmoid curve.

```python
def sigmoid(x):
    return 1/(1+math.e**(-x))
```

```python
x1 = np.linspace(-10, 10, 500) 
y1 = sigmoid(x1)
```

```python
plt.plot(x1, y1)
plt.title('Sigmoid Function')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

## Relu Function :
ReLU, or Rectified Linear Unit, is a common activation function in artificial neural networks defined as \(f(x)=\max (0,x)\). It outputs the input directly if it's positive and returns zero if it's negative.

```python
def relu(x):
    return np.where(x>=0, x, 0)
```

```python
x2 = np.linspace(-10, 20, 500) 
y2 = relu(x2)
```

```python
plt.plot(x2, y2)
plt.title('Relu Function')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

## tanh :
tanh is the hyperbolic tangent function, a mathematical function that maps any real input to a value between -1 and 1. It is often used as an activation function in neural networks because it is zero-centered, helping to speed up training. It is defined as the ratio of the hyperbolic sine to the hyperbolic cosine, or using the exponential function: \(\frac{e^{x}-e^{-x}}{e^{x}+e^{-x}}\). Â 

```python
def tanh(x):
    return (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
```

```python
x3 = np.linspace(-5, 5, 500) 
y3 = tanh(x3)
```

```python
plt.plot(x3, y3)
plt.title('Relu Function')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

## Leaky ReLU
* Leaky ReLU is a modified version of ReLU designed to fix the problem of dead neurons. Instead of returning zero for negative inputs it allows a small, non-zero value. It introduces a slight modification to the standard ReLU by assigning a small, fixed slope to the negative part of the input. This ensures that neurons don't become inactive during training as they can still pass small gradients even when receiving negative values.* 

```python
def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)
```

```python
x4 = np.linspace(-50050, 50050, 500000) 
y4 = leaky_relu(x4)
```

```python
plt.plot(x4, y4)
plt.title('Relu Function')
plt.xlabel('X')

plt.ylabel('Y')
```
